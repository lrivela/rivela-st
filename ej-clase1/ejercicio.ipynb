```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Punto 1

```python
x = np.linspace(0, 10, 100)
noise = np.random.normal(0, 5, 100)

y = x**2
y = y + noise
```

```python
plt.figure(figsize=(10,7))
plt.plot(x, y)
plt.xlabel('Tiempo')
plt.ylabel('Valores')
plt.title('Funcion cuadratica con ruido gaussiano media nula y varianza 5')
```
